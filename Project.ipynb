```python
import pandas as pd
```

```python
df=pd.read_csv('Customer_purchase.csv')
df
```

```python
df.sort_values(by='CustomerName')
df['Revenue']=df['PurchaseQuantity']*df['PurchasePrice']
df
```

```python
#total purchase
total=df['PurchaseQuantity'].sum()
print(total)
```

```python
#total Revenue
total_revenue=df['Revenue'].sum()
print(total_revenue)
```

```python
#average purchase value
avg_purchase=df['Revenue'].mean()
print(avg_purchase)
```

```python
import matplotlib.pyplot as plt
```

```python
top_customers=df.sort_values(by='Revenue',ascending=False)
bars=top_customers.head(10)
fig=plt.figure(figsize=(15,5))
plt.bar(bars['CustomerName'],bars['Revenue'],color='maroon',width=0.3)
plt.xlabel('CustomerName')
plt.ylabel('Revenue')
plt.title('TOP CUSTOMERS AND THEIR PURCHASING BEHAVIOUR')
plt.show()
```

```python
df['Year']=pd.DatetimeIndex(df['purchasedate']).year
df['Quarter']=pd.DatetimeIndex(df['purchasedate']).quarter
df
```

```python
data=df[['Year','Quarter','Revenue']]
data
trend=data.groupby(['Year','Quarter']).sum(['Revenue'])
trend.plot()
```

```python
data_top=df[['ProductCategory','PurchaseQuantity','Revenue']]
graph=data_top.groupby('ProductCategory').sum()
graph.head()
```

```python
label='Electronics','Home Appliances'
plt.pie(graph['Revenue'],labels=label)
plt.show()
```

```python
data_name=df[['ProductName','Revenue']]
graph=data_name.groupby('ProductName').sum()
dts=graph.sort_values(by='Revenue',ascending=False).head()
dts
```
